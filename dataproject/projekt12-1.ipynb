# Data
We import our datasets *time_series_covid19_confirmed_global.csv* and *time_series_covid19_deaths_global.csv* from [Novel Coronavirus (COVID-19) Cases Data](https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases?fbclid=IwAR3CcpqzpkQ9EUCFaR9md0o6S9nCEk2Qb63yLfdtPxnJsOPoDwn8Yuv6llA). We import our datasets as pandas dataframes and exclude the variables Province, Lat, Long that indicate states/provinces, Latitude and Longitude respectively. 

```python
# Get dataset "Confirmed" (Corona) as csv
confirmed = pd.read_csv(r'C:\Users\Runi\Desktop\PythonData\time_series_covid19_confirmed_global.csv')
deaths = pd.read_csv(r'C:\Users\Runi\Desktop\PythonData\time_series_covid19_deaths_global.csv')

# Remove (drop) unnecessary columns for "Confirmed" (Province, Lat, Long)
confirmed = confirmed.drop(confirmed.columns[[0, 2, 3]], axis=1)
deaths = deaths.drop(deaths.columns[[0, 2, 3]], axis=1)
```

## Sum by countries 

In our datasets, some countiries are given more then once since, since there is more then one province in these and each province have their own reportet and death cases of covid-19. By summing we can an overall statistic for each country as a whole. This is done by the groupby() function which groups countries and then the sum method. 

```python
listColumns = list(confirmed.columns)
confirmed1 = confirmed.groupby(['Country/Region'])[listColumns].sum()
confirmed1.head(50) # 50 first 
```

```python
listColdeaths = list(deaths.columns)
deaths1 = deaths.groupby(['Country/Region'])[listColdeaths].sum()
deaths1.head(50)
```

## Creating a variable for Country/Region

Here we create a column Country/Region which is a variable  that contain all countries

```python
# Do Country/Region as a Column again (Named as confirmed2)
confirmed2 = confirmed1.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')
confirmed2.head(10)

deaths2 = deaths1.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')
deaths2.head(10)
```

# Print for specific countries

We now choose the relevant countries for our analysis. These choices are based on different developments of confirmed Covid-19 cases and deaths within the countries and different approaches for tackling the spread of the virus.

```python
confirmed3 = confirmed2.loc[(confirmed2['Country/Region'] == 'China') | (confirmed2['Country/Region'] == 'Denmark') | (confirmed2['Country/Region'] == 'Iran') | (confirmed2['Country/Region'] == 'Italy') | (confirmed2['Country/Region'] == 'Korea, South') | (confirmed2['Country/Region'] == 'US') | (confirmed2['Country/Region'] == 'Spain')]
confirmed3
```

```python
deaths3 = deaths2.loc[(deaths2['Country/Region'] == 'China') | (deaths2['Country/Region'] == 'Denmark') | (deaths2['Country/Region'] == 'Iran') | (deaths2['Country/Region'] == 'Italy') | (deaths2['Country/Region'] == 'Korea, South') | (deaths2['Country/Region'] == 'US') | (deaths2['Country/Region'] == 'Spain')]
deaths3
```

# Show bar-plot for confirmed3 newest date

```python
confirmed3.plot(kind='bar',x='Country/Region',y='3/26/20',color='red')
ax = plt.gca()
ax.set_ylabel('Total number of infected people 3/26/20')
ax.set_title('Total confirmed infected by COVID-19') # Title 
ax.legend().set_visible(False)
plt.show()
```

# Transpose the data AND assign top row as new header

For making time series easier to work with, we transpose the data, so that we get the dates as the row index and change the header-names.

```python
# Transpose the data AND assign top row as new header
Con3_T=pd.DataFrame.transpose(confirmed3)

new_header = Con3_T.iloc[0]
Con3_T= Con3_T[1:]
Con3_T.columns = new_header
Con3_T.head()
Con3_T.index.names = ['Date']
Con3_T.columns.names = ['column']
Con3_T

deaths_T=pd.DataFrame.transpose(deaths3)

new_header = deaths_T.iloc[0]
deaths_T= deaths_T[1:]
deaths_T.columns = new_header
deaths_T.head()
deaths_T.index.names = ['Date']
deaths_T.columns.names = ['column']
deaths_T
```

# Conclusion
We find that there have been some large differences in the development of the Covid-19 virus across countries. 
China and South Korea are known to have taken very quick action, and we can see them breaking the curve of exponential growth, relatively quickly. This is especially telling for China, which has a  very large population and therefore potential for a huge amount of infected. However, we note that some sources do not consider the Chinese data as representative, and speculate that it may be manipulated. Therefore the data from China should be viewed with caution.

The US, Italy and Spain have, on the other hand, been relatively slow to adapt countermeasures, and here we see an explosive growth, which is yet to flatten.
When adjusting for population, we find that China and South Korea have been far more successful than the US, Italy, Spain and Iran at tackling the virus.

Assuming a constant rate of testing, and that data has not been manipulated, countermeasures appear to be successful at slowing the explosive spread of the Covid-19 virus.
